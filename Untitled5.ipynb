```python
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch
```

```python
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
pipe1 = StableDiffusionPipeline.from_pretrained(model_id1,
torch_dtype=torch.float16, USE_SAFETENSORS=True)
pipe1.to("cuda")
```

```python
prompt ="A majestic butterfly howling round a horse, surrounded by a beautiful garden, digital art"
```

```python
image = pipe1(prompt).images[0]
image.save("butterfly.png")
```

```python
plt.imshow(image)
plt.axis('off')
```

```python
def generate_image(pipe,prompt,params):
  img = pipe(prompt,**params).images
  num_images = len(img)
  if num_images > 1:
    fig, ax = plt.subplots(1,num_images,figsize=(20,20))
    for i in range(num_images):
      ax[i].imshow(img[i])
      ax[i].axis('off')
  else:
    fig = plt.figure(figsize=(20,20))
    plt.imshow(img[0])
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
import streamlit as st
from diffusers import StableDiffusionPipeline
import torch
from PIL import Image
model_id = "dreamlike-art/dreamlike-diffusion-1.0"
pipe = StableDiffusionPipeline.from_pretrained(model_id,
torch_dtype=torch.float16, use_safetensors=True)
pipe.to("cuda")
st.title("� AI Image Generator with Stable Diffusion")
prompt = st.text_input("Enter your prompt:", "A majestic butterfly in a garden")
if st.button("Generate Image"):
  with st.spinner("Generating..."):
    image = pipe(prompt).images[0]
    st.image(image, caption="Generated Image",
use_container_width=True)
```
